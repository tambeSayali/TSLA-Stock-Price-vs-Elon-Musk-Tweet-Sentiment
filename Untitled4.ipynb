```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
tweets = pd.read_csv("/content/sentiment_output_fixed.csv")
tsla = pd.read_csv("/content/TSLA.csv")

# Ensure date columns are in datetime format
tweets['Date'] = pd.to_datetime(tweets['Timestamp'])
tsla['Date'] = pd.to_datetime(tsla['Date'])

# Convert likes and retweets to numeric values (e.g., "11.4K" to numeric)
def convert_to_numeric(value):
    value = str(value).replace(',', '').replace('K', 'e3').replace('M', 'e6')
    try:
        return float(eval(value))
    except:
        return np.nan

tweets['Likes'] = tweets['Likes'].apply(convert_to_numeric)
tweets['Retweets'] = tweets['Retweets'].apply(convert_to_numeric)

# Assign numerical values for sentiment (positive=1, neutral=0, negative=-1)
sentiment_mapping = {'positive': 1, 'neutral': 0, 'negative': -1}
tweets['Sentiment_Score'] = tweets['Sentiment'].map(sentiment_mapping)

# Calculate weighted sentiment
tweets['Weighted_Sentiment'] = tweets['Sentiment_Score'] * (tweets['Likes'] + tweets['Retweets'] + 1)

# Aggregate daily sentiment (weighted average)
tweet_summary = tweets.groupby('Date').agg(
    avg_weighted_sentiment=('Weighted_Sentiment', 'mean'),
    total_tweets=('Tweet', 'count')
).reset_index()

# Calculate daily stock price change
tsla['Price_Change'] = tsla['Close'] - tsla['Open']

# Merge the tweet summary with Tesla stock data
merged_data = pd.merge(tsla, tweet_summary, how='left', on='Date')

# Drop rows with missing sentiment or price change values
merged_data = merged_data.dropna(subset=['avg_weighted_sentiment', 'Price_Change'])

# Calculate the correlation
correlation = merged_data['avg_weighted_sentiment'].corr(merged_data['Price_Change'])
print(f"Pearson Correlation: {correlation}")

# Plot the data
plt.figure(figsize=(10, 6))
sns.regplot(
    x='avg_weighted_sentiment',
    y='Price_Change',
    data=merged_data,
    line_kws={'color': 'red'},
    scatter_kws={'alpha': 0.5}
)

# Add labels and title
plt.title("Sentiment vs. Stock Price Change", fontsize=16)
plt.xlabel("Average Weighted Sentiment", fontsize=14)
plt.ylabel("Stock Price Change (Close - Open)", fontsize=14)
plt.grid(alpha=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
tweets = pd.read_csv('sentiment_output_fixed.csv', parse_dates=['Timestamp'])
stocks = pd.read_csv('TSLA.csv', parse_dates=['Date'])

# Preprocess tweets data
tweets['Date'] = tweets['Timestamp'].dt.date  # Extract date from timestamp
tweets['Sentiment_Score'] = tweets['Sentiment'].map({'positive': 1, 'neutral': 0, 'negative': -1})  # Map sentiment to numeric values

# Aggregate daily sentiment scores
daily_sentiments = tweets.groupby('Date')['Sentiment_Score'].mean().reset_index()  # Mean sentiment per day

# Preprocess stock data
stocks['Date'] = stocks['Date'].dt.date  # Ensure date is in correct format
stocks['Daily_Change'] = stocks['Close'].pct_change()  # Calculate daily stock price change

# Merge the datasets on 'Date'
merged = pd.merge(stocks, daily_sentiments, on='Date', how='inner')

# Calculate the correlation matrix between sentiment score and stock price change
correlation = merged[['Daily_Change', 'Sentiment_Score']].corr()

# Output the correlation matrix
print("Correlation Matrix between Daily Change and Sentiment Score:")
print(correlation)

# Visualize the correlation
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap: Stock Price Change and Sentiment Score')
plt.show()

# Optional: Visualize the relationship between sentiment score and stock price change
plt.figure(figsize=(12, 6))
sns.scatterplot(data=merged, x='Sentiment_Score', y='Daily_Change', hue='Sentiment_Score', palette='coolwarm')
plt.title('Sentiment Score vs Stock Price Change')
plt.xlabel('Sentiment Score')
plt.ylabel('Stock Price Change (%)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the datasets
tweets = pd.read_csv('sentiment_output_fixed.csv', parse_dates=['Timestamp'])
stock = pd.read_csv('TSLA.csv', parse_dates=['Date'])

# Map sentiment text to numeric values
sentiment_mapping = {
    'positive': 1,
    'neutral': 0,
    'negative': -1
}
tweets['Sentiment_Score'] = tweets['Sentiment'].map(sentiment_mapping)

# Preprocess the tweet sentiment data
# Group tweets by day and calculate the average sentiment score
tweets['Date'] = tweets['Timestamp'].dt.date  # Extract date from timestamp
daily_sentiment = tweets.groupby('Date')['Sentiment_Score'].mean().reset_index()
daily_sentiment.rename(columns={'Sentiment_Score': 'Average_Sentiment'}, inplace=True)
daily_sentiment['Date'] = pd.to_datetime(daily_sentiment['Date'])

# Merge the sentiment data with stock price data
merged_data = pd.merge(stock, daily_sentiment, left_on='Date', right_on='Date', how='inner')

# Calculate daily percentage change in stock price
merged_data['Price_Change'] = merged_data['Close'].pct_change() * 100

# Drop rows with NaN values (e.g., first row due to pct_change)
merged_data = merged_data.dropna()

# Calculate Pearson correlation coefficient
correlation = merged_data[['Average_Sentiment', 'Price_Change']].corr().iloc[0, 1]
print(f"Correlation between Average Sentiment and Daily Stock Price Change: {correlation:.2f}")

# Plot scatter plot for sentiment vs. stock price change
plt.figure(figsize=(10, 6))
plt.scatter(merged_data['Average_Sentiment'], merged_data['Price_Change'], alpha=0.6, c='blue')
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='red', linestyle='--', linewidth=1)
plt.xlabel("Average Sentiment Score")
plt.ylabel("Daily Stock Price Change (%)")
plt.title("Sentiment vs. Daily Stock Price Change")
plt.show()

# Heatmap to visualize correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(merged_data[['Average_Sentiment', 'Price_Change']].corr(), annot=True, cmap='coolwarm', cbar=True)
plt.title("Correlation Heatmap")
plt.show()

```
